```
from scipy import stats
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
param_dist = {'n_estimators': stats.randint(100, 400),
              'learning_rate': stats.uniform(0.001, 0.5),
              'colsample_bytree': stats.uniform(0.5, 0.9),
              'min_child_weight': [1, 2, 3, 4],
              'max_depth':[50,75,100,150]
             }
tuning= RandomizedSearchCV(classifier,param_distributions=param_dist,n_iter=5,verbose=3)
tuning.fit(X,Y.values.ravel())
tuning.best_estimator_, tuning.best_params_
```
